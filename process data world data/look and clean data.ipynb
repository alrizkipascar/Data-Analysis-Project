```python
import pandas as pd
fertil = 'fertil data.worldbank.xls'
inflation = 'inflation data.worldbank.xls'
```

```python
fertilDF = pd.read_excel( fertil, sheet_name="Data" )
inflationDF = pd.read_excel( inflation, sheet_name="Data" )
```

```python
fertilDF = fertilDF.fillna(0)
inflationDF = inflationDF.fillna(0)
```

```python

gdp = 'worldbank/GDP growth annual Data Worldbank.xls'
gcp = 'worldbank/Gross capital formation (annual % growth) Data Worldbank.xls'
Unemployment = 'worldbank/Unemployment data worldbank.xls'
```

```python
gdpDF = pd.read_excel(gdp, sheet_name="Data" )
gcpDF = pd.read_excel(gcp, sheet_name="Data" )
UnemploymentDF = pd.read_excel(Unemployment, sheet_name="Data" )
#inflationDF = pd.read_excel( inflation, sheet_name="Data" )
```

```python
gdpDF = gdpDF.fillna(0)
gcpDF = gcpDF.fillna(0)
UnemploymentDF = UnemploymentDF.fillna(0)
```

```python
reshaped_gdp = pd.melt(gdpDF, id_vars=["Country Name","Country Code","Indicator Name","Indicator Code"], var_name="Year")
reshaped_gcp = pd.melt(gcpDF, id_vars=["Country Name","Country Code","Indicator Name","Indicator Code"], var_name="Year")
reshaped_Unemployment = pd.melt(UnemploymentDF, id_vars=["Country Name","Country Code","Indicator Name","Indicator Code"], var_name="Year")
```

```python
reshaped_gdp = reshaped_gdp.sort_values(by=["Country Name","Year"],ignore_index=True)
reshaped_gcp = reshaped_gcp.sort_values(by=["Country Name","Year"],ignore_index=True)
reshaped_Unemployment = reshaped_Unemployment.sort_values(by=["Country Name","Year"],ignore_index=True)
```

```python
reshaped_fertil.to_excel("clean_fertil.xlsx")  
reshaped_inflation.to_excel("clean_inflation.xlsx")  
```

```python
reshaped_gdp.to_excel("clean_gdp.xlsx")  
reshaped_gcp.to_excel("clean_gcp.xlsx")  
reshaped_Unemployment.to_excel("clean_Unemployment.xlsx")  
```
